## Getting Started

```python
import pandas as pd
```

```python
closet_data = pd.read_csv('data/clothing.csv')

closet_data.describe()
```

```python
wear_data = pd.read_csv('data/daily-fall-2022.csv')
wear_data.head()

wears = wear_data['clothing'].str.split(pat=",")
clothes = [element for list_ in wears for element in list_]

clothing_counts = pd.Series(clothes).value_counts().to_frame().reset_index()
clothing_counts.columns = ['id', 'wears']
clothing_counts
```

```python
table = pd.merge(closet_data, clothing_counts)
table['cost per wear'] = table['cost'] / table['wears']
```

```python
table['cost per wear'].describe()
```
